Rock, paper & scissor game

```python
### handle the input of user 
def rock_paper_scissor_game():
    from random import choice 
    import time 
    
    while True:
        print("Welcome to Game , Lets see who is Gesture Mastermind!!") 
        trial = 0
        rounds_played = 0
        player_score = 0
        computer_win = 0
        while rounds_played  < 3 :
            #total_rounds_played = 0
            #player_score = 0
            #total_rounds_played = 0
            #computer_win = 0
            
            player_enter = input("Choose your weapon : Rock , Paper or Scissor : ").lower()
            print(f"this is round {rounds_played +1}")
            choose=["paper","scissor","rock"]
            if player_enter.isnumeric():
                print("Enter words only")
                print("Choose Again..")
                continue 
                
            ### player choice and its error 
            
            elif not player_enter in choose:
                print(" Not in weapon list!!  ")
                print(" Choose Again..")
                continue
            ##import random for computer 
            
            print("Now computer turn")
            list_game = ["paper","scissor","rock"]
            #from random import choice 
             
            computer_player = choice(list_game)
            
            ## import waiting time for computer as if its thinking..
           
            print("Computer Chosing......") 
            #import time
            time.sleep(2)
            print(f"Computer : {computer_player.capitalize()}")
            
            if computer_player == player_enter:
                time.sleep(2)
                print("          It's a tie!          ")
                trial += 1
                print(f"This trial : {trial}")
                
               
                continue 
                
            elif computer_player == "rock" and  player_enter == "paper":
                time.sleep(2)
                print("           You won this round !!       ")
                player_score +=1
                print(f"Player score: {player_score}")
                trial += 1
                print(f"This trial : {trial}")
                
                
     
            
            elif computer_player == "paper" and  player_enter == "scissor":
                time.sleep(2)
                print("         You won this round !!       ")
                player_score += 1
                print(f"Player score: {player_score}")
                trial+=1
                print(f"This trial : {trial}")
                

               
                                            
            elif computer_player == "scissor" and  player_enter == "rock":
                time.sleep(2)
                print("           You won this round !!         ")
                player_score += 1
                print(f"Player score: {player_score}")
                trial+=1
                print(f"This trial : {trial}") 
                
        
                                
            else:
                time.sleep(2)
                print("             Computer wins               ")
                computer_win += 1
                print(f"Computer score: {computer_win}")
                trial+=1
                print(f"This trial : {trial}")
                
                
            rounds_played += 1
            
            #print(f"Player score: {player_score}, Computer score: {computer_win}")
               
            print(f" Rounds played : {rounds_played}")
            
          
               
            
            if rounds_played == 3:#tie 
                print(" .....Game Over .....")
                break
        print("And the winner for this match is......\n")
        time.sleep(3)
       
        if player_score > computer_win:
            print(" Gesture Mastermind !!\n You won this match ")
           
            print(f"You won this match in {trial} trials")
        
         
        else:
            print("   Computer.... \n You lost this match  \n!!")    
            ###here im asking if user would like to play again..
        
        print(f"Player score: {player_score}, Computer score: {computer_win}")
        replay = input("Do you want to play again: Y/N")
        if replay.lower() != "y":
            print("      Thanks for playing    \n ")
            break
        
```
